# Missing data problem solved by taking neighbors influence + similar days influence + linear interpolation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_kiosk1 = pd.read_csv('1.csv')
d1 = data_kiosk1.copy()
d1
```

```python
d1.drop(columns=['index','ProductId','KioskId','KioskName','ProductName'],inplace=True)
d1["Date"] = pd.to_datetime(d1["Date"]).dt.date
d1
```

```python
revenue_k1_daily = pd.DataFrame(d1.groupby(['Date'])['GrossPrice'].sum())
revenue_k1_daily.rename(columns={"GrossPrice":"revenue"},inplace=True)
revenue_k1_daily['date'] = pd.to_datetime(revenue_k1_daily.index)
revenue_k1_daily
```

```python
date_range = pd.date_range(start=revenue_k1_daily['date'].min(), end=revenue_k1_daily['date'].max(), freq='D')
complete_data = pd.DataFrame({'date': date_range})
complete_data.info()
```

```python
# Merge the complete date DataFrame with your original data to fill missing dates
data = complete_data.merge(revenue_k1_daily, on='date', how='left')
fig, ax = plt.subplots(1, figsize = (20, 8)) 
ax.grid()
fig.autofmt_xdate()
plt.title("Missing dates from Revenue generated from Kiosk 1 - Cola Mix")
plt.plot(data.date,data.revenue)
plt.show()
```

```python
data[data.revenue.isna()].shape
```

### Using only Interpolation - notice how there are only simple lines to connect the missing gaps

```python
data['revenue'] = data['revenue'].interpolate(method='linear',axis=0)
fig, ax = plt.subplots(1, figsize = (20, 8)) 
ax.grid()
fig.autofmt_xdate()
plt.title("Missing data from Revenue generated from Kiosk 1 - Cola Mix")
plt.plot(revenue_k1_daily.date,revenue_k1_daily.revenue)
plt.show()
```

### Our method 

```python
def impute_revenue(row, data, window_size=30):
    date = row['date']
    missing_revenue = row['revenue']
    
    # Calculate the local mean (a) within the specified window
    local_mean = data[(data['date'] >= date - pd.DateOffset(days=window_size)) & 
                      (data['date'] <= date + pd.DateOffset(days=window_size))]['revenue'].mean()
    
    # Calculate the mean of similar days (b) before the missing date
    similar_days = data[(data['date'] < date) & (data['date'].dt.dayofweek == date.dayofweek)]
    
    # Filter out NaN values from similar days
    similar_days = similar_days.dropna(subset=['revenue']).tail(window_size)
    
    similar_mean = similar_days['revenue'].mean() if not similar_days.empty else data.revenue.mean()
    
    # Impute the missing data using the specified formula
    imputed_value = 0.7 * local_mean + 0.3 * similar_mean
    
    # Replace NaN revenue values with the imputed value
    if pd.isna(missing_revenue):
        return imputed_value
    
    return missing_revenue
```

```python
data = complete_data.merge(revenue_k1_daily, on='date', how='left')
```

```python
# Apply the impute_revenue function to fill missing revenue values
data['revenue'] = data.apply(lambda row: impute_revenue(row, data,5), axis=1)

# interpolate any remaining NaN values
data['revenue'] = data['revenue'].interpolate(method='linear',axis=0)
```

```python
# Sort the DataFrame by 'date' to ensure it's in chronological order
data = data.sort_values(by='date')

# Reset the index
data.reset_index(drop=True, inplace=True)
```

```python
fig, ax = plt.subplots(1, figsize = (20, 8)) 
ax.grid()
fig.autofmt_xdate()
plt.title("Locality influence + similar days influence + linear interpolation - Cola Mix")
plt.plot(data.date,data.revenue)
plt.show()
```

## Testing on data from 2 more kiosks with highest transactions

```python
def return_complete_series(data):
    data.drop(columns=['index','ProductId','KioskId','KioskName','ProductName'],inplace=True)
    data["Date"] = pd.to_datetime(data["Date"]).dt.date
    data = pd.DataFrame(data.groupby(['Date'])['GrossPrice'].sum())
    data.rename(columns={"GrossPrice":"revenue"},inplace=True)
    data['date'] = pd.to_datetime(data.index)
    date_range = pd.date_range(start=data['date'].min(), end=data['date'].max(), freq='D')
    complete_data = pd.DataFrame({'date': date_range})
    data_final = complete_data.merge(data, on='date', how='left')
    return data_final
```

```python
data_kiosk2 = pd.read_csv('2.csv')
data_kiosk2.info()
```

```python
data_kiosk2 = return_complete_series(data_kiosk2)
```

```python
fig, ax = plt.subplots(1, figsize = (20, 8)) 
ax.grid()
fig.autofmt_xdate()
plt.title("Missing dates from Revenue generated from Kiosk 2 - Cola Mix")
plt.plot(data_kiosk2.date,data_kiosk2.revenue)
plt.show()
```

```python
# Apply the impute_revenue function to fill missing revenue values
data_kiosk2['revenue'] = data_kiosk2.apply(lambda row: impute_revenue(row,data,5), axis=1)

# interpolate any remaining NaN values
data_kiosk2['revenue'] = data_kiosk2['revenue'].interpolate(method='linear',axis=0)
fig, ax = plt.subplots(1, figsize = (20, 8)) 
ax.grid()
fig.autofmt_xdate()
plt.title("Locality influence + similar days influence + linear interpolation - Cola Mix")
plt.plot(data_kiosk2.date,data_kiosk2.revenue)
plt.show()
```
