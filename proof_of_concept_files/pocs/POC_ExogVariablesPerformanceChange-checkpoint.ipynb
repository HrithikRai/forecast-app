```python
def impute_revenue(row, data, window_size=30):
    date = row['date']
    missing_revenue = row['revenue']
    
    # Calculate the local mean (a) within the specified window
    local_mean = data[(data['date'] >= date - pd.DateOffset(days=window_size)) & 
                      (data['date'] <= date + pd.DateOffset(days=window_size))]['revenue'].mean()
    
    # Calculate the mean of similar days (b) before the missing date
    similar_days = data[(data['date'] < date) & (data['date'].dt.dayofweek == date.dayofweek)]
    
    # Filter out NaN values from similar days
    similar_days = similar_days.dropna(subset=['revenue']).tail(window_size)
    
    similar_mean = similar_days['revenue'].mean() if not similar_days.empty else data.revenue.mean()
    
    # Impute the missing data using the specified formula
    imputed_value = 0.7 * local_mean + 0.3 * similar_mean
    
    # Replace NaN revenue values with the imputed value
    if pd.isna(missing_revenue):
        return imputed_value
    
    return missing_revenue
```

```python
# Apply the impute_revenue function to fill missing revenue values
data['revenue'] = data.apply(lambda row: impute_revenue(row, data,5), axis=1)

# interpolate any remaining NaN values
data['revenue'] = data['revenue'].interpolate(method='linear',axis=0)
```

```python
# Sort the DataFrame by 'date' to ensure it's in chronological order
data = data.sort_values(by='date')

# Reset the index
data.reset_index(drop=True, inplace=True)
```

### Model 1 = No exogenous variables

```python
# Fine tuning the parameters for the SARIMA based model
data_time_series = data[['date','revenue']]
data_time_series.index = data['date']
data_time_series.drop(['date'],axis=1,inplace=True)
data_time_series.head()
```

```python
df = data_time_series.copy()
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from itertools import product
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
# to evaluate models via a grid search approach

def evaluate_sarima_model(order, s_order, train_data, test_data):
    
    model = SARIMAX(train_data, order=order, seasonal_order=s_order, enforce_stationarity=True, enforce_invertibility=False)
    fitted_model = model.fit(disp=False)
    predicted_values = fitted_model.predict(start=len(train_data), end=len(train_data) + len(test_data) - 1, dynamic=False)
    rmse = sqrt(mean_squared_error(test_data, predicted_values))
    return rmse

# Function to perform grid search for SARIMA parameters

def grid_search_sarima(train_data, test_data, p_values, d_values, q_values, P_values, D_values, Q_values, s_values):
    
    best_rmse = float('inf')
    best_order = None
    best_s_order = None

    for p in p_values:
        for d in d_values:
            for q in q_values:
                for P in P_values:
                    for D in D_values:
                        for Q in Q_values:
                            for s in s_values:
                                order = (p, d, q)
                                s_order = (P, D, Q, s)
                                try:
                                    rmse = evaluate_sarima_model(order, s_order, train_data, test_data)
                                    print(f'SARIMA{order}{s_order} RMSE: {rmse}')
                                    if rmse < best_rmse:
                                        best_rmse = rmse
                                        best_order = order
                                        best_s_order = s_order
                                except Exception as e:
                                    print(f'Error for SARIMA{order}{s_order}: {e}')

    print(f'Best SARIMA Model: SARIMA{best_order}{best_s_order} with RMSE: {best_rmse}')    
    return best_order, best_s_order
```

```python
# Performing train test split

train_size = int(len(df) * 0.8)
train_data, test_data = df[:train_size]['revenue'], df[train_size:]['revenue']
```

```python
train_data
```

```python
train_data.diff().describe()
```

```python
# Ranges of hyperparameters to search
p_values = range(0, 3)
d_values = range(0, 2)
q_values = range(0, 3)
P_values = range(0, 2)
D_values = range(0, 2)
Q_values = range(0, 2)
s_values = [7]  
```

```python
best_order, best_s_order = grid_search_sarima(train_data, test_data, p_values, d_values, q_values, P_values, D_values, Q_values, s_values)
```

```python
best_order
```

```python
best_s_order
```

```python
final_model = SARIMAX(df['revenue'], order=best_order, seasonal_order=best_s_order, enforce_stationarity=False, enforce_invertibility=False)
final_fitted_model = final_model.fit(disp=False)
```

```python
# Make predictions for future periods
future_periods = 25  # adjust as needed
future_forecast = final_fitted_model.get_forecast(steps=future_periods)
```

```python
# Plot the results
ax = df['revenue'].plot(label='Observed', figsize=(20,8))
future_forecast.predicted_mean.plot(ax=ax, label='Forecast', color='red')
ax.set_xlabel('Date')
ax.set_ylabel('Revenue')
plt.title('SARIMA Forecast')
plt.legend()
plt.show()
```
